```python
top_violations = df['Violation Description'].value_counts().head(10).index
df_top = df[df['Violation Description'].isin(top_violations)]

all_boroughs = df['Primary Borough Name'].unique()

plt.figure(figsize=(14,8))
sns.countplot(data=df_top, y='Violation Description', hue='Primary Borough Name', 
              order=top_violations, hue_order=all_boroughs)
plt.xlabel('Count')
plt.title('Top 10 Violation Descriptions by Borough')
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.show()
```

```python
pct_data = df.groupby('Primary Borough Name','BRONX')['Class'].value_counts(normalize=True).mul(100).rename('Percentage').reset_index()

plt.figure(figsize=(10,6))
sns.barplot(data=pct_data, x='Primary Borough Name', y='Percentage', hue='Class')
plt.xlabel('Borough')
plt.ylabel('Percentage of Violations (%)')
plt.title('Violations by Borough and Class (%)')
plt.xticks(rotation=45)
plt.legend(title='Class')
plt.show()
```

```python
df['Inspection Date'] = pd.to_datetime(df['Inspection Date'], errors='coerce')
df['Approved Date'] = pd.to_datetime(df['Approved Date'], errors='coerce')

df['DaysToFix'] = (df['Approved Date'] - df['Inspection Date']).dt.days

print(df['DaysToFix'].describe())
```

```python
# Average Time To Fix Violation 

plt.figure(figsize=(8,5))
sns.barplot(data=df, x='Class', y='DaysToFix', estimator='mean')
plt.xlabel('Violation Class')
plt.ylabel('Average Days to Fix')
plt.title('Average Time to Fix by Violation Class')
plt.show()
```

```python
# Total Violations By Borough
borough_pct = df['Primary Borough Name'].value_counts(normalize=True) * 100

plt.figure(figsize=(10,6))
borough_pct.plot(kind='bar')
plt.xlabel('Borough')
plt.ylabel('Percentage of Violations (%)')
plt.title('Violations by Borough (%)')
plt.xticks(rotation=45)
plt.show()
```

```python
plt.figure(figsize=(10,6))
sns.histplot(data=df, x='Primary Borough Name', hue='Class', multiple='stack',
             shrink=0.8)
plt.xlabel('Borough')
plt.ylabel('Number of Violations')
plt.title('Total Violations by Borough (Stacked by Class)')
plt.xticks(rotation=45)
plt.legend(title='Class A,B,C')
plt.show()
```
